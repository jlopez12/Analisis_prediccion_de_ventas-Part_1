#Inspección de los Datos

```python
# Inspecciona los tipos de datos de cada columna utilizando el atributo dtypes.
print(f"Tipos de datos:\n\n{df.dtypes}")
```

```python
# Cuenta los valores únicos en la columna Producto utilizando el método value_counts().
cnt_valores_unicos_product = df["Product Category"].value_counts()
print(f"Conteo de valores unicos en Producto:\n\n{cnt_valores_unicos_product}")
```

```python
# Muestra todos los valores únicos en la columna Customer ID utilizando el método unique().
cnt_valores_unicos_Customer_id = df["Customer ID"].unique()
print(f"Valores unicos en Customer ID:\n\n{cnt_valores_unicos_Customer_id}")
```

#Filtrado de Datos


```python
# Filtra el DataFrame para mostrar solo las filas donde el Total de ventas Sea mayor a 50.
fltr_total_sales = (df["Total Amount"] > 50)
mayores_50 = df[fltr_total_sales]
print(f"Filas donde el Total Amount es mayor a 50:\n\n{mayores_50}")
```

```python
# Filtra el DataFrame para mostrar solo las filas donde el precio (Precio) sea menor a 0.5.
fltr_price_sales = (df["Price per Unit"] < 0.5)
menores_ = df[fltr_price_sales]
print(f"Filas donde los precios son menor a 0.5:\n\n{menores_}")
```

```python
# Utilizando el método query(), filtra el DataFrame para mostrar las filas donde el producto sea Beauty y las ventas sean mayores a 30.
datos_filtrados = df.query("`Product Category` == 'Beauty' and `Total Amount` > 30")
print(f"Filas donde el Tipo de producto sea Beauty y el total de ventas sea mayor a 30:\n\n {datos_filtrados}")
```

#Slicing de Datos

```python
# Selecciona y muestra solo las columnas Producto y Ventas del DataFrame.
product_total = df[["Product Category","Total Amount"]]
print(f"Categorias y Total de venta:\n\n{product_total}")
```

```python
# Utilizando loc[], selecciona y muestra las filas de la 5 a la 10 (inclusive) y las columnas Producto y Total Amount.
columnas_5_10 = df.loc[5:10, ["Product Category","Total Amount"]]
print(f"Filas desde 5 a 10 mostrando solo Categoria Producto y Total Amount:\n\n {columnas_5_10}")
```

```python
# Utilizando iloc[], selecciona y muestra las primeras 5 filas y las primeras 3 columnas del DataFrame.
primeras_5f_3c = df.iloc[:5, :3]
print(f"Primeras 5 filas y primeras 3 columnas:\n\n{primeras_5f_3c}")
```
